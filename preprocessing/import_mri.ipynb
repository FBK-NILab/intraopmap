```python
import csv
import re
import os
import glob
import numpy as np
```

Load CVS dataset

```python
csv_fd = open(src_csv)
csv_reader = csv.reader(csv_fd)
csv_header = next(csv_reader)
csv_rows = []
for row in csv_reader:
    csv_rows.append(row)
csv_fd.close()
```

```python
sub_max = 308
```

```python
sub_lab = csv_rows[0][1]
folder = "%s/%s" % (src_mri, sub_lab.split('_')[0])
folder
```

```python
for sub in range(sub_max):
    sub_id = csv_rows[sub][0]
    sub_lab = re.sub('è|é','e',csv_rows[sub][1])
    sub_dir = '%s/%s' % (src_mri, sub_lab)
    if not os.path.exists(sub_dir):
        sub_lab = re.sub('_[A-Z,a-z]*$', '', sub_lab)
        sub_dir = '%s/%s' % (src_mri, sub_lab)
    if not os.path.exists(sub_dir):
        sub_lab = re.sub('_[A-Z,a-z]*$', '', sub_lab)
        sub_dir = '%s/%s' % (src_mri, sub_lab)
    if not os.path.exists(sub_dir):
        sub_lab = re.sub('_[A-Z,a-z]*$', '', sub_lab)
        sub_dir = '%s/%s' % (src_mri, sub_lab)
    if not os.path.exists(sub_dir):
        sub_lab = re.sub('_[A-Z,a-z]*$', '', sub_lab)
        sub_dir = '%s/%s' % (src_mri, sub_lab)
    if not os.path.exists(sub_dir):
        print("WARNING %s not found" % csv_rows[sub][1])
    else:
        sub_mri = glob.glob('%s/*nii' % sub_dir)[0]
        sub_dir = '%s/sub-%s/anat' % (out_mri, sub_id)
        cmd = 'mkdir -p %s' % sub_dir
        os.system(cmd)
        sub_t1w = '%s/sub-%s/anat/sub-%s_T1mdc.nii' % (out_mri, sub_id, sub_id)
        cmd = "cp '%s' %s" % (sub_mri, sub_t1w)
        print('sub-%s_T1mdc.nii' % sub_id)
    #os.system(cmd)
```
