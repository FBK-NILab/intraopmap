```python
# Ubuntu
FIX_ENV = 'DISPLAY=:0'
CLOUD_COMPARE = '/snap/bin/cloudcompare.CloudCompare'
```

```python
# MacOS
FIX_ENV = 'DISPLAY=:0'
CLOUD_COMPARE = '/Applications/CloudCompare.app/Contents/MacOS/CloudCompare'
```

```python
import os
import csv
```

```python
def load_csv(src_csv):
    csv_fd = open(src_csv)
    csv_reader = csv.reader(csv_fd)
    csv_header = next(csv_reader)
    csv_rows = []
    for row in csv_reader:
        csv_rows.append(row)
    csv_fd.close()
    return csv_header, csv_rows
```

```python
def split_3d_points_by_roi(hdr, row, out, tag='cortical'):
    roi_split = []
    rois = list(set([r[1] for r in row]))
    xyz = {}
    for roi in rois: 
        xyz[roi]=[]
    for r in row:
        xyz[r[1]].append([r[2],r[3],r[4],r[5]])
    for roi in rois:
        roi_split.append(os.path.join(out, 'des-%s_roi-%s.csv' % (tag, roi)))
        with open(roi_split[-1], 'w') as out_csv:
            for des in xyz[roi]:
                out_csv.write(",".join(des) + "\n")
    return roi_split
```

```python
def csv2ply(csv):
    cmd = "%s %s -SILENT -AUTO_SAVE OFF -C_EXPORT_FMT PLY -O \'%s\' -NO_TIMESTAMP -SAVE_CLOUDS" % \
        (FIX_ENV, CLOUD_COMPARE, csv)
    os.system(cmd)
```

```python
def des_3d_ply(src_csv, out_dir='', tag='cortical_batch-03'):
    if out_dir == '':
        out_dir = os.path.dirname(src_csv)
    hdr, row = load_csv(src_csv)
    roi_split = split_3d_points_by_roi(hdr, row, out_dir, tag=tag)
    for roi in roi_split:
        csv2ply(roi)
        os.system("rm %s" % roi)
```

Split CSV file with DES 3D points by ROI and convert them to PLY format

```python
des_3d_ply(src_dir + '/des-cortical_batch-03.csv')
```
